# The RC Circuit Model of a neuron

### In our last tutorial we wrapped up describing the passive properties of neurons, that is, the properties of neurons at rest. Specifically, we showed that the current across the cellular membrane for a particular ion can be given by the following equation, which we derived from Ohms law
$$ \Large I_X = (V_M - V_X)g_X $$
### where $I_X, V_X$ and $g_X$ are the current, equilibrium/Nernst potential and channel conducatance for a particular ionic species $X$ and $V_M$ is the membrane voltage. What we did not mention last time, is that the this equation we discussed above is exacly the same equation you would use to describe the current through an RC electrical circuit, the kind of thing you might see in a physics or electrical engineering course. They are, for all intents and purposes, equivalent circuits. So what is an RC circuit?<break>

### An RC circuit has two components, a resistor and a capacitor. Thuogh you might not realize it, you are actually already familiar with resistors. Resistors limit the current that can flow through a conductor, similarly to how a bottleneck in a pipe can limit the volume of water that can flow through it. In our circuit model of a neuron, our resistors are our ion channels, and their resistance is simply the inverse of their conductance, and vice versa.

$$\Large g_X = \frac{1}{R_X} $$

In fact, the canonical form of writing Ohms law is really $V = IR$, but biophysicists typically think of things in terms of conductances rather than resistances.

### So if resistors are the equivalent of ion channels in our model, what is the capacitor analagous to, and what is it doing?

 *“Like other cells, neurons are packed with a huge number and variety of ions and molecules. A cubic micron of cytoplasm might contain, for example, 1010 water molecules, 108 ions, 107 small molecules such as amino acids and nucleotides, and 105 proteins. Many of these molecules carry charges, either positive or negative. Most of the time, there is an excess concentration of negative charge inside a neuron. Excess charges that are mobile, like ions, repel each other and build up on the inside surface of the cell membrane. Electrostatic forces attract an equal density of positive ions from the extracellular medium to the outside surface of the membrane.*

*The cell membrane is a lipid bilayer 3 to 4 nm thick that is essentially impermeable to most charged molecules. This insulating feature causes the cell membrane to act as a capacitor by separating the charges lying along its interior and exterior surfaces”*

**Theoretical Neuroscience, Dayan and Abbott**

### In an electrical circuit, a capacitor acts identically to the cellular membrane. It is typically two metal plates separated by an dielectric medium/insulator, which, when current is injected into the system, has charge of opposite signs build up on either plate.

### In the above image we can see how the RC circuit analogy works more diagramatically. The green rectangles here represents the potassium channel, with its unique conducatence $g_K$. In circuit terms we think of this as a resistor with a unique resistance and we represent it with the zig-zag line. The the black bars on the left with charge accumulated on either side of them represents a capacitor with its capacitance $C$, which is a property of the membrane. Finally we represent the driving force of potassium with the symbol for a battery with its unique voltage, the Nernst potential for potassium $V_K$.

### Okay, so we can analogize the electrical behavior of the cell to that of an RC circuit, how does that help us to *understand* what is going on within a cell?

### Imagine the cell were to experience a sudden infusion of current. Maybe the cell received a signal from another neuron or we are in lab injecting current artificially, how does the circuit react? The first thing to know is that when this happens, the total current across the membrane $I_M$ can be divided into two parts, the current caused by ions flowing through ion channels $I_R$ (the resistive pathway), and the current caused by a buildup of charge on the inside of the membrane $I_C$ (the capacitive part). The former of these we've already discussed, it is just given by Ohms law, but for the second we have to introduce an equation describing the current caused by a capacitor

$$\Large I_C = C_M\frac{dV_M(t)}{dt} $$

### Now, if you have not taken calculus the fraction on the right hand side of the equation may look strange or intimidating, but you can think about it as simply the rate of change of the membrane potential as a function of time. So the current through the capacitor is just capacitance of the membrane (which we will assume to be a constant) times the rate of change of the membrane potential. It is really okay if that doesn't fully click right now, but with that equation, we can now describe the current flow across the membrane like so:
$$\Large I_M = I_R + I_C = g_KV_m + C_m \frac{dV_M}{dt}$$

### Now we have a differential equation that we can solve for $V_M$ to get how the membrane potential will react to our injected current $I_M$. If you want to know the maths behind actually isolating $V_M$ I recommend you look into taking Math 207, or better yet the AMath equivalent. Nonetheless, at the end of our process we are left we out answer:

$$\Large V_M(t) = I_M(t) R_K(1 - e^{\frac{-t}{RC}}) $$
