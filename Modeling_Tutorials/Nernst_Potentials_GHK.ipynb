# Membrane Potential and the Nernst Equation

### In our class discussions, we've emphasized the crucial functions of both leaky and gated ion channels in maintaining the neuron's membrane potential at a resting value. Leaky ion channels allow a small, but steady flow of ions in or out of the cell, whilst gated ion channels open in response to specific stimuli such as changes in voltage or binding of neurotransmitters. The opening of these gated channels facilitate very rapid changes in the membrane potential. These rapid changes in potential are very important and form the basis for cellular communication across the brain and body, but today we are going to discuss the resting potential, and we can derive it quite naturally using ideas from statistical physics. To start, there are two things that I need you to take as given.<break>

### 1) The average thermal energy of a single ion is approximately $k_BT$ where $ k_B$ is the Boltzman constant $(1.38 \times 10^{-23}\frac{m^2kg}{s^2K})$ and $T$ is the absolute temperature in Kelvin. Since we usually deal with moles, we usually multiply this quantity by Avogadro's number, $N_Ak_BT = RT$, where $R$ is the universal gas constant $(8.31 \frac{J}{mol K^\circ})$.

### 2) The energy gained or lost when a mole of ions crosses a membrane with a potential difference $V_T$ equals $FV_T$ where F is the Faraday constant $(96,480 \frac{\text{Coloumbs}}{mol})$

### Equating these two terms, we see that for the average thermal energy of a mole of ions to equal the energy associated with moving that mole of ions across a membrane, the membrane voltage must equal the following ratio

$$\Large V_T = \frac{RT}{F} $$

### This is an important relation between chemical and electrical forces that will more easily allow us to uncover the equilibrium potential of a cell.

### As was discussed in class the electrochemical gradient is the difference in electrochemical potential between the inside and outside of the cell. As its name suggests, the electrochemical gradient has two contributions: the first of these is the chemical force which is is driven by differences in ion concentrations across the membrane; the second is the electrical force, which is driven by the electrical potential across the membrane. As we know from chemistry and physics, particles and molecules are driven to diffuse and equalize their concentration across barriers, and particles of the same or opposite charge repel or attract one another respectively. We can summarize the effects of these two forces via the Nernst equation. The Nernst equation gives us the equilibrium potential, the point at which the electrical forces and the chemical forces are balanced such that there is no *net* movement of ions across the membrane. When the membrane potential is equal to the Nernst potential, there is *no net movement of ions across the membrane*. Ions may still be moving across the membrane, but their net effect on the membrane potential is zero.

### If an ion has electrical charge $zq$ where $z$ is the valence and $q$ is the charge of one proton, then its thermal energy must be at least $-zqV$ to cross the membrane (remember $V$ has units of $\frac{J}{\text{Coloumb}}$). So what is the probability that an ion, or a mole of ions has energy greater than $-zqV$? This isn't a statistical mechanics course so we will take it as fact that the probability in molar units is equal to $\exp(\frac{zFV}{RT}) = \exp(\frac{zV}{V_T})$. Because this quantity is a probability, we can say that the concentration of the ions inside the cell with at least this energy is equal to $[\text{inside}]\exp(\frac{zV}{V_T})$ and therefore at equilibrium, when the concentration of ions outside the cell with the energy to cross the membrane is equal to the concentration of ions inside the cell with the energy to cross the membrane, the following equation must be satisfied.

$$\Large [\text{outside}] =  [\text{inside}]\exp\Biggl(\frac{zV}{V_T}\Biggr)$$

### Rearranging for our particular $V$ we arrive at the Nernst equation.

$$\Large V_{ion} = \frac{RT}{zF}\ln\Biggl(\frac{[\text{ion}]_{\text{outside}}}{[\text{ion}]_{\text{inside}}}\Biggr) $$


### In the Nernst equation, $R, F$ and $T$ correspond to the gas constant, Faradays constant and the temperature in Kelvin. For us neuroscientists, the temperature does not change much so we usually simplify the Nernst equation to 
$$\Large V_{ion} = \frac{62mV}{z}\log\Biggl(\frac{[\text{ion}]_{\text{outside}}}{[\text{ion}]_{\text{inside}}}\Biggr) $$

### Where $z$ is the valence of the ion, and the base of our logarithm is now base 10 rather than base $e$. <break>
### So what are these equilibrium potentials and how do they contribute to the resting potential of a neuron? For our sake lets consider just two ions, potassium $\text{K}^+$ and sodium $\text{Na}^+$ . If you feel confused so far, I suggest looking at the Nernst equation examples from class before looking at the following plots.

### Lets do potassium first, at rest, the concentration of potassium is ~10x greater inside the cell than outside. For our purposes we'll say $[\text{K}^+]_{\text{in}}= 150mM$ and $[\text{K}^+]_{\text{out}}= 15mM$. Therefore, we have the following Nernst potential for potassium.
$$\Large V_{\text{K}^+} = \frac{62mV}{+1}\log\Biggl(\frac{15}{150}\Biggr) = 62mV \cdot -1 = -62mV $$
### What does this number mean? The fact that $V_{\text{K}^+} = -62mV$ means that when the electric potential inside the cell is equal to $-62mV$, the rate of $\text{K}^+$ moving in will equal the rate of $\text{K}^+$ moving out of the cell. This should hopefully make an intuitive sense, since we need a large attractive electrical force inside the cell to offset the large concentration difference driving $\text{K}^+$ out of the cell. Lets run through an example.<break>

**1) Lets say that we start with an initial membrane voltage of $-50mV$. a number more positive than the $V_\text{K}^+$ we just calculated. Both the concentration gradient and the electrical gradient causes $\text{K}^+$ to leave the cell. As it does, positive potassium ions are leaving so the membrane voltage becomes more negative, slowing the efflux of $\text{K}^+$ until eventually $V_m = V_\text{K} = -62mV$ and the ion is at equilibrium.**


### For $\text{Na}^+$ I will spare us the math, but know that the concentration of $\text{Na}^+$ is higher *outside* the cell rather than inside. This should be enough to tell you that the equilibrium potential for $\text{Na}^+$ is a positive number. In fact it is ~$+55mV$. Given that number, try to think about what forces are interacting on it and where it would flow if given the opportunity.

### Okay, so we have two equilibrium potentials, one for potassium and one for sodium, and they seem to be pointing in opposite directions. That is, sodium wants to flow *into* the cell, and potassium wants to flow *out*. How do these forces interact to actually determine the membrane potential? To figure this out we have to introduce one extremely important relationship, 

# **Ohms Law**<break>

### Ohms Law is simply a mathematical equation which states that the electric current between two points through some conductor is directly proportional to the electrical potential across the two points. More formally it can be written as 

$$\Large Vg = I $$

### Where $I$ is electrical current, $g$ is conductance, and $V$ is the electrical potential. Unlike in electronics, the charged particles carrying this current are ions like potassium, sodium, and chloride, rather than electrons. The conductance is our proportionality constant and quantifies the proportion of channels that are open and permeable to a particular ion. For now, for the sake of simplicity we will think of channel conductances as unchanging.

### If we limit ourselves to one ion and one channel then we can quantify the current carried by that ion through that channel by the product of the deviation of the membrane potential from that ions equilibrium potential and the conductance for its channel. Lets take potassium for example. From our equation, we have that
$$\Large I_{\text{K}^+} = (V_m - V_{\text{K}^+})g_{\text{K}^+} $$

### In words, the current through the potassium channel is proportional to the deviation of the membrane potential from the potassium equilibrium potential, and the proportionality constant is exactly the conductance of the potassium channel. We call the difference between the membrane potential $V_m$ and an ions equilibrium potential (in our case this is $V_{\text{K}^+})$ its **driving force**.

# Plotting
### Lets take play around with this idea visually and see what current we get for different inter- and extra-cellular concentrations of potassium, as well as channel conductances. Press the play button on the cell below to generate an interactive plot and feel free to play with the sliders.

Note that in neuroscience convention, negative current is defined as electrical charge flowing *into* the cell, whilst positive current is defined as electrical charge exiting the cell. That is why the electrical current axis appears inverted in the plot above.<break>

Also note that sometimes if you move the sliders too quickly, the image may appear to glitch. Usually you can just press the play button again and it should resolve itself.

```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interactive

# Function to calculate V_k based on K_out and K_in
def calculate_V_k(K_out, K_in):
    return 62 * np.log10(K_out / K_in)  # Nernst equation for potassium

# Prepare the initial plot
fig, ax = plt.subplots(figsize=(10, 4))

# Move left y-axis and bottom x-axis to centre, passing through (0,0)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

# Show ticks in the left and lower axes only


V_m_range = np.linspace(-100, 100, 400)  # Membrane potential for x-axis
delta_I_line, = ax.plot(V_m_range, np.zeros_like(V_m_range))  # Initial line
ax.set_ylim(15, -15)  # Adjust the y-axis limits if necessary
ax.set_xlabel(r'$V_m (mV)$')
ax.set_ylabel(r'$I_{K^+}$')
ax.xaxis.set_ticks([-100, -75, -50, -25, 25, 50, 75, 100], [-100, -75, -50, -25, 25, 50, 75, 100])
ax.yaxis.set_ticks([-15, -12.5, -10, -7.5, -5, -2.5, 2.5, 5, 7.5, 10, 12.5, 15], [-15, -12.5, -10, -7.5, -5, -2.5, 2.5, 5, 7.5, 10, 12.5, 15])
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_label_coords(1, .55)  # x label positioned at end of the axis
ax.yaxis.set_label_coords(0.5, 1.1)

# Function to update the plot
def update(K_out=15, K_in=150, g_k=0.1):
    V_k = calculate_V_k(K_out, K_in)
    delta_I = (V_m_range - V_k) * g_k  # ΔI calculated over the range of V_m
    delta_I_line.set_ydata(delta_I)
    fig.canvas.draw_idle()

# Create the interactive sliders for K_out, K_in, and g_k
interactive_plot = interactive(update, 
                               K_out=(1.0, 150.0), 
                               K_in=(1.0, 150.0), 
                               g_k=(0.0, 1.0, 0.01))
display(interactive_plot)
```

### What we've just done is construct a single ion model of the resting potential. Where the above line crosses zero *is* our resting potential, and since we have only included a single ion, potassium, $V_\text{rest} = V_{\text{K}^+}$. In other words, the resting potential *is* the equilibrium potential of potassium. The points on the line tell tell us the direction and magnitude of potassium current driving the cell back to $V_{\text{K}^+}$ when $V_m \neq V_{\text{K}^+}$. <break>

# Two-Ion Models
### But what about other ions that we know are present in and around the cell like sodium? How do we incorporate them into this model? What is the new resting potential? Let's start from first principles...<break>

### If we know that at the resting potential the net current in and out of the cell is zero than we can confidently say that the current contributions from each ion must also sum to zero. After all, that is what it means to be at the resting potential.

$$\Large I_\text{Na} + I_\text{K} = 0 $$

### Now using Ohms law, we can replace our currents with the product of their respective driving forces and channel conductances.

$$\Large (V_\text{rest}- V_\text{Na})g_\text{Na} + (V_\text{rest} - V_\text{K})g_\text{K} = 0$$

### Now lets do some algebra and isolate $V_\text{rest}$

$$\Large V_\text{rest}g_\text{Na} - V_\text{Na}g_\text{Na} + V_\text{rest} g_\text{K} -  V_\text{K}g_\text{K} = 0 $$

$$\Large V_\text{rest}g_\text{Na} + V_\text{rest} g_\text{K} = V_\text{Na}g_\text{Na} + V_\text{K}g_\text{K}$$

$$\Large V_\text{rest}(g_\text{Na} + g_\text{K}) = V_\text{Na}g_\text{Na} + V_\text{K}g_\text{K} $$

$$\Large V_\text{rest} = \frac{V_\text{Na}g_\text{Na} + V_\text{K}g_\text{K}}{g_\text{Na} + g_\text{K}} $$

### What have we just shown? We have constructed a two ion model of the resting potential and shown from first principles, that the resting potential is a ***weighted average*** of each individual ions equilibrium potential, where the weights are their respective channel conductances. Try and convince yourself that this is intuitive and makes a certain kind of sense. Both ions want to reach their equilibrium potential, and their ability to do so is proportional to the number of channels that are available to facilitate their movement.<break>
### Now lets see how this changes our plot. Press play on the cell again and mess around with the sliders.

```python
%matplotlib widget
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interactive

# Constants
R = 8.314 # J/(mol·K)
F = 96485 # C/mol
T = 298 # K (25 degrees Celsius)

# Function to calculate V_k and V_na based on ion concentrations outside and inside
def calculate_V_ion(c_out, c_in):
    return 62 * np.log10(c_out / c_in)

def find_zero_crossings(y_values, x_values):
    # Finds the x value where y crosses zero
    zero_crossings = np.where(np.diff(np.sign(y_values)))[0]
    return x_values[zero_crossings]

# Prepare the initial plot
fig, ax = plt.subplots(figsize=(10, 4))

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

# Eliminate upper and right axes
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

V_m_range = np.linspace(-100, 100, 400)  # Membrane potential for x-axis
delta_I_lineK, = ax.plot(V_m_range, np.zeros_like(V_m_range), label=r'$K^+$')  # Initial line
delta_I_lineNa, = ax.plot(V_m_range, np.zeros_like(V_m_range), label=r'$Na^+$')
delta_I_lineRest, = ax.plot(V_m_range, np.zeros_like(V_m_range), label='total')

dot_lineK, = ax.plot([], [], 'bo', label=r'$V_K$')  # Initial dot for K+
dot_lineNa, = ax.plot([], [], 'o', color='orange', label=r'$V_{Na}$')  # Initial dot for Na+
dot_lineRest, = ax.plot([], [], 'go', label=r'$V_{rest}$')  # Initial dot for total current



ax.set_ylim(15, -15)  # Adjust the y-axis limits if necessary
ax.set_xlabel(r'$V_m (mV)$')
ax.set_ylabel(r'$I$')
ax.xaxis.set_ticks([-100, -75, -50, -25, 25, 50, 75, 100], [-100, -75, -50, -25, 25, 50, 75, 100])
ax.yaxis.set_ticks([-15, -12.5, -10, -7.5, -5, -2.5, 2.5, 5, 7.5, 10, 12.5, 15], [-15, -12.5, -10, -7.5, -5, -2.5, 2.5, 5, 7.5, 10, 12.5, 15])
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.xaxis.set_label_coords(1, .55)  # x label positioned at end of the axis
ax.yaxis.set_label_coords(0.5, 1.1)

# Function to update the plot
def update(K_out=4, K_in=140, Na_out=145, Na_in=12, g_k=0.001, g_Na=0.001):
    # Calculate the equilibrium potentials for K+ and Na+
    V_k = calculate_V_ion(K_out, K_in)
    V_na = calculate_V_ion(Na_out, Na_in)

    # Calculate the resting potential as a weighted average
    

    # Calculate the potassium and sodium currents across the membrane potential range
    delta_I_k = g_k * (V_m_range - V_k)
    delta_I_na = g_Na * (V_m_range - V_na)
    V_rest = delta_I_k + delta_I_na
    delta_I_lineK.set_ydata(delta_I_k)
    delta_I_lineNa.set_ydata(delta_I_na)
    delta_I_lineRest.set_ydata(V_rest)
    
    zero_crossings_k = find_zero_crossings(delta_I_k, V_m_range)
    zero_crossings_na = find_zero_crossings(delta_I_na, V_m_range)
    zero_crossings_rest = find_zero_crossings(V_rest, V_m_range)

    # Clear previous points
    dot_lineK.set_data([], [])
    dot_lineNa.set_data([], [])
    dot_lineRest.set_data([], [])
    if len(zero_crossings_k) > 0:
        dot_lineK.set_data(zero_crossings_k[0], 0)
    if len(zero_crossings_na) > 0:
        dot_lineNa.set_data(zero_crossings_na[0], 0)
    if len(zero_crossings_rest) > 0:
        dot_lineRest.set_data(zero_crossings_rest[0], 0)
    
    fig.canvas.draw_idle()
    fig.legend()

interactive_plot = interactive(update,
                               K_out=(1.0, 150.0),
                               K_in=(1.0, 150.0),
                               g_k=(0.01, 1.0, 0.01),
                               Na_out=(1.0, 150.0),
                               Na_in=(1.0, 150.0),
                               g_Na=(0.01, 1.0, 0.01))
display(interactive_plot)
```
